### Missing Values

Kita dapat gunakan fungsi `dropna()` untuk menghapus data yang tidak memiliki nilai dan fungsi `fillna()` untuk mengisi nilai yang kosong

```python
df.isna()
```

```python
df.notna()
```

```python
# count all missing values
df.isna().sum()
```

- Missing value di kolom `Age` akan diisi dengan rata-rata dari semua umur penumpang (menggunakan mean)
- Missing value di kolom `Cabin` akan dihapus
- Missing value di kolom `Embarked` akan diisi dengan nilai Embarked yang paling banyak muncul di dataset (menggunakan modus)

```python
# Age
# Langkah 1
rata_umur = df['Age'].mean()
# Langkah 2
df['Age'] = df['Age'].fillna(rata_umur)
# Langkah 3
df['Age'].isna().sum()
```

```python
# Cabin
df.drop('Cabin', axis=1, inplace=True)
```

```python
# Embarked
# Langkah 1
modus = df['Embarked'].mode()[0]
# Langkah 2
df['Embarked'] = df['Embarked'].fillna(modus)
# Langkah 3
df['Embarked'].isna().sum()
```

```python
# cek lagi
df.isna().sum()
```

## map and filter

- Fungsi `map()` hanya bekerja pada Pandas Series
- Fungsi `apply()` dapat bekerja pada Pandas Series dan DataFrame

```python
# case membuat dataframe
data = {
    'nama' : ["Anton","Budi","Caca","Deni","Eflin"],
    'tugas': [9.0,8.8,6.6,7.8,9.6],
    'uts' : [72,52,32,92,72],
    'uas': [93,66,99,73,97]}

df_nilai = pd.DataFrame(data)

df_nilai.head()
```

```python
type(df)
```

```python
df_nilai.sort_values(by=['uas'], ascending=False)
```

```python
# kolom tugas * 10 karena dalam skla 1-10
df_nilai['tugas'].map(lambda x : x * 10)
```

```python
df_nilai['tugas'] = df_nilai['tugas'].map(lambda x : x * 10)
```

```python
df_nilai
```

```python
# buat nilai akhir dengan prosentase tugas 30%, uts 30%, uas 40%
```

```python
# cara 1
df_nilai[['tugas','uts','uas']].apply(lambda col : (col['tugas']*0.3 + col['uts']*0.3 + col['uas']*0.4), axis=1)
```

```python
df_nilai['final'] = df_nilai[['tugas','uts','uas']].apply(lambda col : (col['tugas']*0.3 + col['uts']*0.3 + col['uas']*0.4), axis=1)
```

```python
df_nilai
```

```python
# cara 2
def hitung(col):
    return col['tugas']*0.3 + col['uts']*0.3 + col['uas']*0.4
```

```python
df_nilai['total'] = df_nilai[['tugas','uts','uas']].apply(hitung, axis=1)
df_nilai
```
