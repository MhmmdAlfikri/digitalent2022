## Python Env



Cara install Virtual Env

- Jalankan `pip install virtualenv` melalui CMD / Terminal
- Buat Folder `mkdir projekku` dan masuk ke folder tersebut `cd projekku`
- Buat virtual env `virtualenv env-baru`
- Aktifkan virtual env di linux `source env-baru/bin/activate` atau di windows `env-baru\Scripts\activate.bat`
- Untuk keluar dari virtual env `deactivate`
