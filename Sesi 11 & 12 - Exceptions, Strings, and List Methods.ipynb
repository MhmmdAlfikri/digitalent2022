## Exception

exceptions can be handled using a `try .. except ..` statement.

```python
# initialize the amount variable
nilai = 10000

# perform division with 0
a = nilai / 0
print(a)
```

```python
try:
    nilai = 10000

    # perform division with 0
    a = nilai / 0
    print(a)
except Exception as e:
    print(e)
```

```python
try:
    i = int("Hello!")
except Exception as e:
    print(e)
```

```python
try:
    print(a+2)
except Exception as e:
    print(e)
```

```python
try:
    f = open("demofile.txt", "r")
    print(f.read())
except Exception as e:
    print(e)
```

### `try` dengan klausa `else`

Dalam beberapa situasi, ketika ingin menjalankan blok kode tertentu jika blok kode di dalam `try` berjalan tanpa kesalahan. Untuk kasus ini, Gunakan `else` dengan pernyataan try.

```python
try:
    angka = int(input("Enter a number: "))
except:
    print("Not an even number!")
else:
    angka = angka * 2
    print(angka)
```

### `try` dengan klausa `finally`

statement di `finally` dijalankan walau dalam kondisi apapun

```python
try:
    f = open("demofile.txt", "r")
    print(f.read())
except Exception as e:
    print(e)
finally:
    print("Cleaning up the mess")
```

```python
kalimat = "Tadi saya pergi ke pasar. Di Pasar Saya ketemua banyak orang"

newlist = []

with open("stopword.txt", "r") as file:
    word = file.read().split(" ")

    for i in kalimat.split(" "):
        if i.lower() in word:
            continue
            
        newlist.append(i)
        
print(newlist)
newkalimat = ' '.join(newlist)
print(newkalimat)
```

## Lab

- Buat game sederhana dengan menebak angka yang akan keluar secara random dari range 1-5
- jika input angka 10 (atau lainnya) maka keluar program 
- Jika input angka diluar 1-5 kasih info salah input dan lakukan input ulang 
- Hasil tebakan dicetak diakhir loop di list dengan urutan `[n, m] ->` dimana `n adalah benar dan m adalah salah`

#### Contoh output
- masukkan angka : 7
- `salah input`
- masukkan angka : 3
- salah, random 2 | tebakan anda 3
- `benar 0, salah 1`

- masukkan angka : 3
- salah, random 5 | tebakan anda 3
- `benar 0, salah 2`

- masukkan angka : 10
- `exit program`

```python
from random import randint


status = [0,0]
while True:
    try:
        tebakan = int(input("masukkan angka : "))
        nilairandom = randint(1,5)    
    
        if tebakan == 10:
            print("keluar program")
            break
        elif tebakan not in [1,2,3,4,5]:
            print("input harus di angka 1,2,3,4,5")
        elif tebakan == nilairandom:
            status[0] += 1
            print("benar, nilai {} | tebakan anda {}".format(nilairandom,tebakan))
        elif tebakan != randint(1,5):
            status[1] += 1
            print("salah, nilai {} | tebakan anda {}".format(nilairandom,tebakan))

        print("benar {}, salah {}".format(status[0],status[1]))
    except:
        print("input harus numeric")
```
